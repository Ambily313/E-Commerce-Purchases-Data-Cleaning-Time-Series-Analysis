## Missing Values Handling

```python
#  check for missing values.
df.isnull().sum()
```

```python
#Count how many rows have at least one missing value.

df.isnull().any(axis=1).sum()
```

```python
# Drop all rows with any missing values.
df_dropna_rows=df.dropna(how='any',axis=0)
df_dropna_rows.sample(3)
```

```python
# Drop columns with all missing values.
df_dropna_col=df.dropna(how='all',axis=1)
df_dropna_col.sample(3)
```

```python
# Fill missing numeric values with the median.
df_fillnum_median = df.copy()
numeric_cols = df_fillnum_median.select_dtypes(include='number').columns
df_fillnum_median[numeric_cols] = df_fillnum_median[numeric_cols].fillna(df_fillnum_median[numeric_cols].median())
```

```python
# Fill missing string values with the mode.
df_fillstr_mode = df.copy()
string_cols = df_fillstr_mode.select_dtypes(include='object').columns
for col in string_cols:
    mode_val = df_fillstr_mode[col].mode()[0]
    df_fillstr_mode[col].fillna(mode_val, inplace=True)
```

```python
# Forward fill missing values.
df_ffill=df.fillna(method='ffill')
```

```python
# Backward fill missing values.

df_bfill = df.fillna(method='bfill')
```

```python
# Interpolate missing numeric values.
df_interp = df.copy()
df_interp[numeric_cols] = df_interp[numeric_cols].interpolate()

```

```python
# Identify columns with the highest % of missing data.

df_missing_percent=df.isnull().mean().sort_values(ascending=False) * 100
df_missing_percent
```

###  Time Series Handling

```python
# Convert Purchase Time to pandas datetime.

# Simulating a dummy time column using AM/PM and a fixed hour
df['Purchase Time'] = pd.to_datetime(df['AM or PM'] + ' 12:00', format='%p %I:%M')

df[['Purchase Time']]
```

%p = AM or PM

%I = Hour (12-hour format)

%M = Minute

```python
# Extract the hour of purchase.

df['Hour'] = df['Purchase Time'].dt.hour

df.sample(5)
```

```python
# Extract the day of the week from the timestamp.
df['Day of Week'] = df['Purchase Time'].dt.day_name()
```

```python
# Find how many purchases were made in the evening (after 6PM).
evening_purchases = df[df['Hour'] < 12]
evening_purchases.shape[0]
```

```python

# Find purchases made on weekends.
weekend_purchases = df[df['Day of Week'].isin(['Saturday', 'Sunday'])]
```

```python
# Create a new column Purchase Month.

df['Purchase Month'] = df['Purchase Time'].dt.month_name()
```

```python
# Find the earliest and latest purchase dates.
earliest = df['Purchase Time'].min()
latest = df['Purchase Time'].max()

print(earliest)
print(latest)
```

```python
# Calculate time difference between the first and last purchase.
time_diff = latest - earliest

time_diff
```

```python
# Group purchases by day.
purchases_by_day = df.groupby(df['Purchase Time'].dt.date).size()
```

### String Operations

```python
# Get unique email domains (like gmail.com).
df['domains']=df['Email'].apply(lambda x:x.split('@')[1])
unique_domain=df['domains'].unique()
unique_domain
```

```python
# Find how many users have “.edu” emails.
edu_count = df[df['Email'].str.contains('.edu')].shape[0]
edu_count
```

```python
# Filter rows where Job Title contains “engineer”.
engineers=df[df['Job'].str.contains('engineer',case=False)]
engineers
```

```python
# Create a new column: IsSenior → True if job title contains “Senior”.

df['IsSenior']=df['Job'].str.contains('Senior',case=False)

df[df['IsSenior']==True].shape[0]
```

```python
# Extract the username part of the email (before '@').

df['Email Username'] = df['Email'].apply(lambda x: x.split('@')[0])
df['Email Username']
```

```python
# Count how many users use Yahoo as email provider.
yahoo_users = df[df['domains'].str.contains('yahoo.com', case=False)].shape[0]
yahoo_users
```

```python
# Convert all job titles to lowercase.

df['Job Lowercase'] = df['Job'].str.lower()
df.sample(3)
```

```python
# Replace spaces in job titles with underscores.

df['Job Underscore'] = df['Job'].str.replace(' ', '_')
df.sample(3)
```

```python
#Check if all emails are valid (contain '@').
df['Valid Email'] = df['Email'].apply(lambda x: '@' in x)
valid_email_check = df['Valid Email'].all()
valid_email_check
```

```python
#Create a column EmailDomainLength.
df['EmailDomainLength'] = df['domains'].apply(len)
df.sample(5)
```

```python
# View data types of all columns.
# df.dtypes
df.info()
```

```python
# Convert boolean columns (if any) to int.
bool_cols=df.select_dtypes(include='bool').columns
for i in bool_cols:
    df[i]=df[i].astype(int)



df.info()
```

```python
# Change CC Exp Date column to datetime (MM/YY format)
df['CC Exp Date']=pd.to_datetime(df['CC Exp Date'], format='%m%y')
```

```python
# Split CC Exp Date into Exp_Month and Exp_Year.

df['Exp_month']=df['CC Exp Date'].dt.month_name
df['Exp_year']=df['CC Exp Date'].dt.year
```

```python
# Create column CardType → 'Visa', 'MasterCard', etc. from first digit.

def card_type(x):
    first_digit=str(x)[0]
    if first_digit == '3':
        return 'American Express'
    elif first_digit == '4':
        return 'Visa'
    elif first_digit == '5':
        return 'MasterCard'
    elif first_digit == '6':
        return 'Discover'
    else:
        return 'Other'
    
df['card_type']=df['Credit Card'].astype(str).apply(card_type)

df[['card_type']]
```
